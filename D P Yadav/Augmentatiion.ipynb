```python
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import tkinter as tk
from PIL import ImageTk, Image
import cv2
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img


import os
        
batch_size = 1

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
        rescale=1./255,
                                             horizontal_flip=True,
                                             vertical_flip=True,
                                             rotation_range=30,
                                              zoom_range=[1, 1.5], 
                                             #width_shift_range=0.3,
                                             #height_shift_range=0.3,
                                              shear_range=0.3
                                             #channel_shift_range=0.3
)
 

# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(rescale=1./255)

# this is a generator that will read pictures found in
# subfolers of 'data/train', and indefinitely generate
# batches of augmented image data
train_generator = train_datagen.flow_from_directory(
         'C:\\Users\\Glau\\Desktop\\DP_Covid\\',  # this is the target directory
        target_size=(256,256),  # all images will be resized to 150x150
        batch_size=batch_size,
        class_mode=None,shuffle=True)  # since we use binary_crossentropy loss, we need binary label
print(len(train_generator))
aug_img=[next(train_generator)[0] for i in range(20000)]
data=[]
for i in aug_img:
    data.append(array_to_img(i))
print(data[0])
```
